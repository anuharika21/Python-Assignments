### Question 1: Write a Python program to count the number of elements in a list without using len.


- `sample input : `
    - `var = [10,11,12,13]`
- `sample output :`
    - `result : 4`

```python
# Please write The solution Here for above question 1:
var=[10,11,12,13]
count=0
for i in var:
  count+=1
print(count)

```

### Question 2: Write a Python program to find the maximum and minimum values in a given list.

- `sample input : `
    - `var = [10,11,12,13,14,15]`
- `sample output :`
    - `result : max = 15 , min = 10`

```python
# Please write The solution Here for above question 2:
var=[10,11,12,13,14,15]
max=var[0]
min=var[0]

for i in var:
  if i>max:
    max=i
  if i<min:
    min=i
print("max=",max,",min=",min)

```

### Question 3: Write a Python program to remove the first occurrence of a specified element from a list.

- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : [1,3,4,2,5]`
    

```python
# Please write The solution Here for above question 3:
a = [1,2,3,4,2,5]
d = []
target =2
c =1
for i in a:
  if i == target:
    if c == 1:
      c = c+1
    else:
      d.append(i)
  else:
    d.append(i)
print(d)
```

### Question 4: Write a Python program to find the average of all the elements in a list.


- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : 2.833 `

```python
# Please write The solution Here for above question 4:
var = [1,2,3,4,2,5]
count=0
sum=0
for i in var:
  sum+=i
  count+=1
avg=sum/count
print(f'{avg:.3f}')
```

### Question 6:  Write a Python program to find the index of a specified element in a list..

- `sample input : `
    - `var = [1,2,3,4,2,5]`
- `sample output :`
    - `result : 3 index is 2`

```python
var = [1, 2, 3, 4, 2, 5]

for i in range(len(var)):
    if var[i] == 3:
        print(var[i], "index is", i)
```

### Question 7: Write a Python program to find which number is palindrome in below given list .

- `sample input : `
    - `var = [134,121,155]`
- `sample output :`
    - `result : 134 is not a palindome , 121 is a palindrome `

```python
# Please write The solution Here for above question 7:
var = [134, 121, 155]
s= []

for i in var:

    if str(i) == str(i)[::-1]:
        s.append(f"{i} is a palindrome")
    else:
        s.append(f"{i} is not a palindrome")

print(", ".join(s))
```

### Question 8: Write a Python program to find the common elements between two lists.

- `sample input : `
    - `var1 = [1,2,3,4,10]`
    - `var2 = [10,11,12,14]`
- `sample output :`
    - `result : 10`

```python
# Please write The solution Here for above question 8:
var1 = [1,2,3,4,10]
var2 = [10,11,12,14]
common=[]
for i in var1:
  if i in var2:
    common.append(i)
print(common)
```

### Question 9: Write a Python program to remove the duplicates from a list

- `sample input : `
    - `var = [10,11,12,10,12,24]`
- `sample output :`
    - `result : [10,11,12,24]`

```python
# Please write The solution Here for above question 9:
var = [10,11,12,10,12,24]
dup=[]
for i in var:
  if i not in dup:
    dup.append(i)
print(dup)

```
